```python
# Selenium 및 필요한 라이브러리 임포트
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# 크롬 옵션 설정
options = webdriver.ChromeOptions()
options.add_argument("--ignore-local-proxy")

# 드라이버 설정
try:
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)
except:
    driver = webdriver.Chrome()

# 네이버 블로그 검색 페이지로 이동
search_url = "https://search.naver.com/search.naver?ssc=tab.blog.all&query=%EC%95%BC%EA%B5%AC%EC%9E%A5%20%EB%B0%A9%EB%AC%B8%20%ED%9B%84%EA%B8%B0&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20181127to20191127"
driver.get(search_url)

# 페이지 로딩 대기
time.sleep(3)

# 무한 스크롤을 위한 설정
SCROLL_PAUSE_TIME = 2  # 스크롤 후 대기 시간

# 현재 페이지 높이 저장
last_height = driver.execute_script("return document.body.scrollHeight")

# 블로그 링크 저장 리스트
blog_links = set()

while True:
    # 페이지 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 스크롤 후 페이지 로딩 대기
    time.sleep(SCROLL_PAUSE_TIME)

    # 새로운 페이지 높이 계산 후 확인
    new_height = driver.execute_script("return document.body.scrollHeight")

    # 페이지 높이에 변화가 없으면 스크롤 종료
    if new_height == last_height:
        print("더 이상 새로운 게시물이 없습니다.")
        break

    last_height = new_height

    # 현재 페이지에서 블로그 링크 추출
    try:
        links = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a.title_link"))
        )
        
        for link in links:
            href = link.get_attribute('href')
            blog_links.add(href)  # 중복 방지를 위해 set 사용

        print(f"현재까지 추출된 블로그 링크 수: {len(blog_links)}")

    except Exception as e:
        print(f"링크 추출 중 오류 발생: {e}")

# 드라이버 종료
driver.quit()

# 최종 결과 출력
print("\n총 추출된 블로그 링크 수:", len(blog_links))
for i, link in enumerate(blog_links, 1):
    print(f"{i}: {link}")
```
